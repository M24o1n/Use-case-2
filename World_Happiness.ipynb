We will go and check outliers 

From the box plots below, we saw that the score column has no outliers. However we have potential outliers in the other fields.
We will use the z-score and the IQR methods to further investigate the outliers

First the IQR method. We will create a method called IQR


```python
def iqr(df, col):
    col_Q1 = df[col].quantile(0.25)
    col_Q3 = df[col].quantile(0.75)
    col_IQR = col_Q3 - col_Q1

    lower_bound = col_Q1 - 1.5 * col_IQR
    upper_bound = col_Q3 + 1.5 * col_IQR

    df['outlier'] = (df[col] < lower_bound) | (df[col] > upper_bound)
```

Next we will create a simliar function as the above, but using z-score


```python
def z (df, col):
    df['z_outlier'] = np.abs(zscore(df[col])) > 3
```

Now we will apply the IQR and z-score methods to the data


```python
outlier_data = []

for col in df_all.columns:
    if df_all[col].dtype == 'float64':
        iqr(df_all, col)  
        outlier_data.append(df_all[df_all['outlier'] == True])

        z(df_all, col)
        outlier_data.append(df_all[df_all['outlier'] == True])

df_outliers = pd.concat(outlier_data)
```

Then we will consider only the outliers that are both detected by the IQR and z-score methods


```python
df_outliers = df_outliers[df_outliers['outlier'] & df_outliers['z_outlier']]
```

Next question to solve, Did any country experience a significant increase or decrease in happiness?

To answer this question we have to calculate the difference in the score for each country between the years.

We will fill each year column with the difference between its year and the previous year.

```python
df_diff = df_all.pivot(index='Country', columns='Year', values='Score').reset_index()
df_diff['2016'] = df_diff[2016] - df_diff[2015]
df_diff['2017'] = df_diff[2017] - df_diff[2016]
df_diff['2018'] = df_diff[2018] - df_diff[2017]
df_diff['2019'] = df_diff[2019] - df_diff[2018]
df_diff = df_diff[['Country', '2016', '2017', '2018', '2019']]
df_diff.head()

```

We notice that Angola has a null value in 2019. However as we focus on the increase/decrease per year, this will not affect the answer of the question.

```python
# Next we will plot the top 5 countries with the highest increase and the top 5 countries with the highest decrease for each year

fig, axes = plt.subplots(4, 2, figsize=(15, 20))

for i, year in enumerate(['2016', '2017', '2018', '2019']):
    sns.barplot(x=year, y='Country', data=df_diff.nlargest(5, year), ax=axes[i, 0])
    axes[i, 0].set_title(f'Top 5 Countries with the highest increase in {year}')
    sns.barplot(x=year, y='Country', data=df_diff.nsmallest(5, year), ax=axes[i, 1])
    axes[i, 1].set_title(f'Top 5 Countries with the highest decrease in {year}')
plt.tight_layout()
plt.show()
```
